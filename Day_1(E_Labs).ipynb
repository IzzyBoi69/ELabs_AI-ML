```python
xval= list(range(1,9))
y_val1=[x**2 for x in xval]
y_val2=[x**3 for x in xval]
print(xval)
print(y_val1)
print(y_val2)
plt.title("Exponentials")
plt.xlabel("A")
plt.ylabel("B")
plt.tick_params(axis="both", labelsize = 15)
plt.scatter(xval, y_val1,s= 150, c="Green")
plt.scatter(xval, y_val2,s= 150, c="Yellow")
plt.plot(xval,y_val2, c="blue", linewidth= 2)
plt.show()
```
